<!-- # Central Limit Theorem Demonstration -->

The objective of this section is to observe and understand the Central Limit Theorem in action. The CLT is a fundamental result that supports many statistical techniques and methods. It provides a theoretical basis for making inferences about population parameters based on sample statistics.

## Distributions Selection

1. **Uniform Distribution**
2. **Exponential Distribution**
3. **Poisson Distribution**

For each of the distributions, we perform the following steps:

1. Generate a large number of random samples with a specific sample size from the chosen distribution.
2. Calculate the mean of each sample.
3. Plot the histogram of the sample means and overlay it with the expected normal distribution based on the Central Limit Theorem.
4. Repeat steps a to c for increasing sample sizes and observe how the distribution of sample means changes as we increase the sample size.

Now, let's proceed with demonstrating the Central Limit Theorem using the selected distributions.


### Import Needed Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

### Define Parameters for All Distributions

```python
num_samples = 1000  # Number of samples to generate
sample_sizes = [10, 50, 100]
population_size = 10000
population_mean = 0.5  
population_std = 0.1  # Population standard deviation
```

```python
def calculate_sample_means(sample, sample_size):
    return np.mean(sample[:, :sample_size], axis=1)
```

## Uniform Distribution

A uniform distribution is characterized by constant probability density.

```python
# Step 1: Generate random samples
uniform_samples = np.random.uniform(0, 1, (num_samples, population_size))

# Step 2 & 3: Calculate sample means and plot histograms
def plot_histogram_uniform_distribution(sample_means, sample_size, population_mean, population_std):
    plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='g')

    # Plot expected normal distribution
    mu = population_mean
    sigma = population_std / np.sqrt(sample_size)
    x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
    plt.plot(x, norm.pdf(x, mu, sigma), color='r')

    plt.title(f'Sample Size: {sample_size}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')

plt.figure(figsize=(15, 5))
for i, sample_size in enumerate(sample_sizes, 1):
    sample_means = calculate_sample_means(uniform_samples, sample_size)
    # print(sample_means)
    plt.subplot(1, len(sample_sizes), i)
    plot_histogram_uniform_distribution(sample_means, sample_size, population_mean, population_std)

plt.tight_layout()
plt.show()
```

## Exponential Distribution

An exponential distribution is often used to model the time until an event occurs.

```python
# Step 1: Generate random samples
exponential_samples = np.random.exponential(scale=1.0, size=(num_samples, population_size))

def plot_histogram_exponential_distribution(sample_means, sample_size, population_mean, population_std):
    plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='b')

    # Plot expected normal distribution
    mu = 1.0
    sigma = 1.0 / np.sqrt(sample_size)
    x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
    plt.plot(x, norm.pdf(x, mu, sigma), color='r')

    plt.title(f'Sample Size: {sample_size}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')

plt.figure(figsize=(15, 5))
for i, sample_size in enumerate(sample_sizes, 1):
    sample_means = calculate_sample_means(exponential_samples, sample_size)
    plt.subplot(1, len(sample_sizes), i)
    plot_histogram_exponential_distribution(sample_means, sample_size, population_mean, population_std)

plt.tight_layout()
plt.show()
```

## Poisson Distribution

A Poisson distribution is often used to model the number of events occurring in a fixed interval of time or space.

```python
# Step 1: Generate random samples
poisson_samples = np.random.poisson(lam=2.0, size=(num_samples, population_size))

def plot_histogram_poisson_distribution(sample_means, sample_size, population_mean, population_std):
    plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='purple')

    # Plot expected normal distribution
    mu = 2.0
    sigma = np.sqrt(2.0 / sample_size)
    x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
    plt.plot(x, norm.pdf(x, mu, sigma), color='r')

    plt.title(f'Sample Size: {sample_size}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')

plt.figure(figsize=(15, 5))
for i, sample_size in enumerate(sample_sizes, 1):
    sample_means = calculate_sample_means(poisson_samples, sample_size)
    plt.subplot(1, len(sample_sizes), i)
    plot_histogram_poisson_distribution(sample_means, sample_size, population_mean, population_std)

plt.tight_layout()
plt.show()
```
